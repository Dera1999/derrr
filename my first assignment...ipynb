## playing card class.  Task

```python
class card(object):
    ''' a playing card string... mostly useful for doing many things'''
    suits = {'d':'diamonds' , 'h': 'hearts', 's': 'spades'}
    value = {1: 'Ace' 10:'joker'}
    def __init__(self, suits, value):
        self.value = value
        self.suits = suits
    def get_value(self):
        ''' returns value'''
        return self.value
    def get_suits(self):
        ''' returnd the suits'''
        return self.suits
    def __str__(self):
        '''what should be displayed any time this guy is called'''
        my_card = str(self.value) + ' of ' + str(self.suits)
        return my_card
```

## put it in a function so that you can call it anytime..


```python
def Random_walk(num_of_steps):
    steps = [1, -1]
    walk = []
    first_step = choice([1,-1])
    walk.append(first_step)
    for i in range (num_of_steps):
        next_step = walk[i] + np.random.choice([1,-1])
        walk.append(next_step)
    plt.rc('figure', figsize= (12,6))
    plt.plot(walk, c= 'green', marker='+', linestyle= ':' )
    plt.title('The Random Walk Exercise')
    plt.xlabel(' the number of steps')
    plt.ylabel('the acutal movements')
```

```python
def Plot_many(num, steps):
    list_title = list(range(1,num+1)) # no necessarily needed i just made this so i can name my legend with it.
    x = list(range(1, steps + 1))  # creating an x variable
    for i in range(num): # making iterations the number of times inputed
    
        
    
        plt.plot(x, Random_walk(steps), label = 'Plot Number' + ' ' + str(list_title[i]))
        plt.title('Our Random Walk')
        plt.xlabel(' Number of Steps')
        plt.ylabel(' Distance from Origin')
        plt.legend(loc = 'lower left')
    plt.show()   
```

# The sierpinski Triangle

##  the triangular plot of randomness making a whole lot of sense to me...

```python
def trans_1(p):
    x = p[0]
    y = p[1]
    
    x1= 0.5 * x
    y1 = 0.5 * y
    return x1,y1

def trans_2(p):
    x = p[0]
    y = p[1]
    
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1, y1

def trans_3(p):
    x = p[0]
    y = p[1]
    
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1, y1


transformation = [trans_1,trans_2,trans_3]

a1 = [0]
b1 = [0]

a,b = 0,0

for i in range(1000000):
    trans = choice(transformation)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
    
```

```python
def balanced_backet():
    strings = input(' input the brackets let us check : ')
    stack = []
    
    brackets = { '}':'{', ')':'(', ']':'['}
    
    for char in strings:
        if not stack:
            stack.append(char)
        elif char not in brackets:
            stack.append(char)
            print('this is char', char, 'this is stack', stack)
        elif brackets[char] == stack[-1]:
            stack.pop()
            print('This is char', char, 'this is stack', stack)
        else:
            stack.append(char)
        
    if not stack:
        return True
    else:
        return False
    
            
            
            
```

# the towers of hanoi

## this guy is scaling at a rate of (2^n - 1) 

### when moving 3 disks from A to C, we move the top two disks to B and then the bottom disk to C. More generally, when moving n disks from A to C we move the n-1 disks to B: we then move the bask disk from A to C and then move the stack of n-1 on B to c.

```python
# A = [3,2,1]
# B = []
# C = []

count = 0
def towers_of_hanoi(A,B,C,n):
    global count
    
    A = [i for i in range(1, n + 1)]
    A.reverse()
    B = []
    C = []

    
    
    
    if n == 1:
        disk = A.pop()
        C.append(disk)
        count += 1
    else:
        towers_of_hanoi(A,C,B,n-1)
        towers_of_hanoi(A,B,C,1)
        towers_of_hanoi(B,A,C,n-1)
    return count
```
